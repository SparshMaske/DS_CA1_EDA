```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/sparshmaske/Desktop/globalterrorismdb_2021Jan-June_1222dist.csv')
```

```python
print("First 5 rows of the dataset:")
print(df.head())
```

```python
print("\nDataset Info:")
df.info()
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(12, 6))
top_countries = df['country_txt'].value_counts().nlargest(10)
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.title('Top 10 Countries with Most Terrorist Attacks (Jan-June 2021)')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_countries.png')
```

```python
plt.figure(figsize=(12, 6))
region_counts = df['region_txt'].value_counts()
sns.barplot(y=region_counts.index, x=region_counts.values, palette='plasma', orient='h')
plt.title('Terrorist Attacks by Region (Jan-June 2021)')
plt.xlabel('Number of Attacks')
plt.ylabel('Region')
plt.tight_layout()
plt.savefig('attacks_by_region.png')
```

```python
plt.figure(figsize=(12, 6))
attack_counts = df['attacktype1_txt'].value_counts()
sns.barplot(y=attack_counts.index, x=attack_counts.values, palette='magma', orient='h')
plt.title('Most Common Attack Types (Jan-June 2021)')
plt.xlabel('Number of Attacks')
plt.ylabel('Attack Type')
plt.tight_layout()
plt.savefig('common_attack_types.png')
```

```python
plt.figure(figsize=(12, 6))
weapon_counts = df['weaptype1_txt'].value_counts()
sns.barplot(y=weapon_counts.index, x=weapon_counts.values, palette='cividis', orient='h')
plt.title('Most Common Weapon Types (Jan-June 2021)')
plt.xlabel('Number of Attacks')
plt.ylabel('Weapon Type')
plt.tight_layout()
plt.savefig('common_weapon_types.png')
```

```python
plt.figure(figsize=(12, 8))
target_counts = df['targtype1_txt'].value_counts().nlargest(15)
sns.barplot(y=target_counts.index, x=target_counts.values, palette='inferno', orient='h')
plt.title('Top 15 Most Common Target Types (Jan-June 2021)')
plt.xlabel('Number of Attacks')
plt.ylabel('Target Type')
plt.tight_layout()
plt.savefig('common_target_types.png')
```

```python
plt.figure(figsize=(10, 5))
monthly_counts = df['imonth'].value_counts().sort_index()
sns.lineplot(x=monthly_counts.index, y=monthly_counts.values, marker='o')
plt.title('Monthly Trend of Terrorist Attacks (Jan-June 2021)')
plt.xlabel('Month')
plt.ylabel('Number of Attacks')
plt.xticks(range(1, 7), ['January', 'February', 'March', 'April', 'May', 'June'])
plt.grid(True)
plt.savefig('monthly_attack_trends.png')
```

```python
plt.figure(figsize=(12, 6))
```

```python
top_groups = df[df['gname'] != 'Unknown']['gname'].value_counts().nlargest(10)
sns.barplot(y=top_groups.index, x=top_groups.values, palette='rocket', orient='h')
plt.title('Top 10 Terrorist Groups by Number of Attacks (Jan-June 2021)')
plt.xlabel('Number of Attacks')
plt.ylabel('Terrorist Group')
plt.tight_layout()
plt.savefig('top_terrorist_groups.png')
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_cols.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.savefig('correlation_matrix.png')
```

```python
region_attack_cross = pd.crosstab(df['region_txt'], df['attacktype1_txt'])
region_attack_cross.plot(kind='bar', stacked=True, figsize=(14, 8),colormap='viridis')
plt.title('Distribution of Attack Types by Region')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('region_attack_distribution.png')
```

```python
top_groups = df[df['gname'] != 'Unknown']['gname'].value_counts().nlargest(5).index
```

```python
df_top_groups = df[df['gname'].isin(top_groups)]
```

```python
group_attack_cross = pd.crosstab(df_top_groups['gname'], df_top_groups['attacktype1_txt'])
```

```python
group_attack_cross.plot(kind='barh', stacked=True, figsize=(12, 8), colormap='plasma')
plt.title('Preferred Attack Types of Top 5 Terrorist Groups')
plt.xlabel('Number of Attacks')
plt.ylabel('Terrorist Group')
plt.tight_layout()
plt.savefig('group_attack_preferences.png')
```
